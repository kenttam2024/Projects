```r
#Find stores performance in 2012 Q3
library(zoo)
retaildata1 <- retaildata
retaildata1$Quarter<- as.yearqtr(retaildata$Date,format="%d-%m-%Y")
retaildata1$Quarter
Q3 <- filter(retaildata1,Quarter=="2012 Q3")
Q3_performance<- summarise(group_by(Q3,Store),Sales=sum(Weekly_Sales))
arrange(Q3_performance,desc(Sales)) #top Stores performance in Q3
```

```r
#Find holiday has negative impact on sales
non_holiday <- filter(retaildata1,retaildata1$Holiday_Flag=="0")
mean(non_holiday$Weekly_Sales)
holiday <- filter(retaildata1,retaildata1$Holiday_Flag=="1")
mean(holiday$Weekly_Sales) #holiday has higher mean than non-holiday
min(holiday$Weekly_Sales) 
holiday_performance<- summarise(group_by(holiday,Date),Result=sum(Weekly_Sales))
arrange(holiday_performance,Result) #31-12-2021 has the lowest weekly sales
```

```r
#Building multiple regression model 
#data manipulation to filter out "Date" since we added new_date1 

data5 <- select(retaildata,-(Date))
data5
str(data5)
```

```r
#Import library for set seed for random number
library(caTools)
#Set seed to "1" 
set.seed(1)
#Split training dataset 70:30
sample <- sample.split(data5$Weekly_Sales,SplitRatio=0.70)
sample
#train_date is subset of sample that is true
train_data <- subset(data5,sample==TRUE) 
#test_data is subset of sample that is false
test_data <- subset(data5,sample==FALSE)
```

```r
#building model# 1 using train_data
model1 <- lm(Weekly_Sales~., data=train_data)
summary(model1)
```

```r
#final model# 2 with only all significant variables
model2 <- lm(Weekly_Sales~Store+Holiday_Flag+Temperature+CPI+Unemployment , data=train_data)
summary(model2)
```

```r
#predict using test_data
predtest<-predict(model2,test_data)
predtest1<- data.frame(predtest)
predtest1

final_data<- cbind(test_data,predtest1)
final_data <- arrange(final_data,desc(predtest1))
final_data
```

```r
#sum of root mean square
sqrt(mean((final_data$Weekly_Sales - final_data$predtest)^2))
```

```r
#export final_data to .csv file
write.csv(final_data,"/kaggle/working/project1.csv")
```
